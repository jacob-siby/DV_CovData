```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('covid_19_india.csv',encoding='cp1252')
```

```python
data.head()
```

```python
data.head()
```

```python
fig, xy = plt.subplots()  # Create a figure containing a single axes.
xy.plot([1, 2, 4, 5], [1, 2, 6, 5])  # Plot some data on the axes.
```

```python
x=data['Date']
y=data['Deaths']
```

```python
plt.bar(x, y, color = 'r', width = 1, label = "DEATH")
#plt.subplot(2, 100, 1)
plt.xlabel('Date')
plt.ylabel('No: of death')
plt.title('Deaths by Covid19')
plt.legend()
plt.show()
```

```python
x=data['State/UnionTerritory']
y=data['Deaths']
plt.bar(x, y, color = 'r', width = 1, label = "DEATH")
#plt.subplot(2, 100, 1)
plt.xlabel('State/UnionTerritory')
plt.ylabel('No: of death')
plt.title('Deaths by Covid19')
plt.legend()

plt.show()
```
