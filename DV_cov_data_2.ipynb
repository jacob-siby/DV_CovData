#Function
unique values in a list

```python
new_data=pd.read_csv('covid_19_indiaN.csv',encoding='cp1252')

x=new_data['State/UnionTerritory']
def unique(list1):
 
    # intilize a null list
    unique_list = []
     
    # traverse for all elements
    for x in list1:
        # check if exists in unique_list or not
        if x not in unique_list:
            unique_list.append(x)
    # print list
    for x in unique_list:
        print(x)
    return unique_list
U_state=unique(x)
```

```python
print(len(U_state))
```

unique states

```python
for a in range(len(U_state)):
    print(U_state[a])
```

#State Wise Total Count for Pie

```python
d_count=new_data['Daily Count']
d=[]  #Total number of cases per state
s=0   # Total number of cases
for z in range(len(U_state)):
  d.append(0)
  for y in range(len(x)):
   if U_state[z]==x[y]:
    d[z]=d[z]+d_count[y]
    s=s+d_count[y]
  print(U_state[z],d[z])
print(s)


```

```python
x=d
y=U_state


fig = plt.figure(figsize =(35, 35))
plt.pie(x, labels = y)
plt.title('State/UnionTerritory COVID19 Share', fontsize=30, fontweight='bold')

plt.legend()


plt.show()
plt.savefig('State_ConCase_scatter.png')
```

```python
d_data={}
d_data['date']=u_date
d_data['count']=d



x=d_data['date']
y=d_data['count']
  

# plotting a line plot after changing it's width and height
p = plt.figure()
p.set_figwidth(100)
p.set_figheight(70)


plt.plot(x, y)


plt.title('Date VS Confirmed Cases', fontsize=65, fontweight='bold')
plt.xlabel('Date', fontsize=40)
plt.ylabel('Confirmed Cases', fontsize=40)


plt.yticks(fontsize=30)
plt.xticks(fontsize=10, rotation='vertical')
plt.legend() 


plt.show()
p.savefig('Date_CaseCurve_plot.png')
```
